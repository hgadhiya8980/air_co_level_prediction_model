```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("AirQualityUCI.csv")
df.shape
```

```python
df.head()
```

```python
l=list(df["CO_level"].unique())
type(l)
print(l,type(l))
```

```python
for num,var in enumerate(l):
    num+=1
    df["CO_level"].replace(var, num, inplace=True)
```

```python
da=[]
for var in df["Date"]:
    try:
        temp = var.split("/")
        da.append(int(temp[1]))
    except:
        da.append(np.nan)
```

```python
df1=df.join(pd.DataFrame({"date":da}))
df1.head()
```

```python
da1=[]
for var in df["Time"]:
    try:
        temp = var.split(":")
        temp1=(int(temp[0])+int(temp[1])+int(temp[2]))/3
        da1.append(temp1)
    except:
        da1.append(np.nan)
```

```python
df2=df1.join(pd.DataFrame({"time":da1}))
df2.head()
```

```python
df3 = df2.drop(["Date","Time"], axis=1)
df3.head()
```

```python
df3.columns
```

```python
df3_cols=df3.columns
df3_cols
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3.corr(), annot=True)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
for var in df3_cols:
    plt.hist(df3[var], rwidth=0.8)
    plt.title(f"hist of {var}")
    plt.xlabel(f"{var}")
    plt.ylabel("Score")
    plt.show()
```

```python
for var in df3_cols:
    sns.distplot(df3[var])
    plt.title(f"hist of {var}")
    plt.show()
```

```python
for var in df3_cols:
    sns.displot(df3[var])
    plt.title(f"hist of {var}")
    plt.show()
```

```python
for var in df3_cols:
    sns.boxplot(df3[var])
    plt.title(f"hist of {var}")
    plt.show()
```
